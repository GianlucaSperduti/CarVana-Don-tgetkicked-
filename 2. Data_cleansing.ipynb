### MMRAcquisition variables

```python
temp=train[train["MMRAcquisitionAuctionAveragePrice"].isnull()][['Make','SubModel','Model','Trim','Size','WheelType','VehicleAge']]
temp2=train.loc[train['Make'].isin(temp['Make'])]
temp2['MMRAcquisitionAuctionAveragePrice'].groupby(
    [temp2['Make']]).mean()
```

```python
temp2['MMRAcquisitionAuctionAveragePrice'].groupby(
    [temp2['Make'], temp2['SubModel'], temp2['Model']]).mean()
```

```python
def fun_temp(x):
    return x.fillna(x.mean())
```

```python
train['MMRAcquisitionAuctionAveragePrice']=train['MMRAcquisitionAuctionAveragePrice'].groupby(
    [train['Make'], train['SubModel'], train['Model']]).apply(fun_temp)
train['MMRAcquisitionAuctionCleanPrice']=train['MMRAcquisitionAuctionCleanPrice'].groupby(
    [train['Make'], train['SubModel'], train['Model']]).apply(fun_temp)
train['MMRAcquisitionRetailAveragePrice']=train['MMRAcquisitionRetailAveragePrice'].groupby(
    [train['Make'], train['SubModel'], train['Model']]).apply(fun_temp)
train['MMRAcquisitonRetailCleanPrice']=train['MMRAcquisitonRetailCleanPrice'].groupby(
    [train['Make'], train['SubModel'], train['Model']]).apply(fun_temp)
```

```python
train.isnull().sum() #vediamo quanti null ci sono in ogni attr
sum_null=train.isnull().sum()
sum_null.index
        
for name, num_na in zip(sum_null.index, train.isnull().sum()):
    if num_na>0:
        print("{}: {}".format(name,num_na))
```

```python
train['MMRAcquisitionAuctionAveragePrice']=train['MMRAcquisitionAuctionAveragePrice'].groupby(
    [train['Make'], train['SubModel']]).apply(fun_temp)
train['MMRAcquisitionAuctionCleanPrice']=train['MMRAcquisitionAuctionCleanPrice'].groupby(
    [train['Make'], train['SubModel']]).apply(fun_temp)
train['MMRAcquisitionRetailAveragePrice']=train['MMRAcquisitionRetailAveragePrice'].groupby(
    [train['Make'], train['SubModel']]).apply(fun_temp)
train['MMRAcquisitonRetailCleanPrice']=train['MMRAcquisitonRetailCleanPrice'].groupby(
    [train['Make'], train['SubModel']]).apply(fun_temp)
```

```python
train.isnull().sum() #vediamo quanti null ci sono in ogni attr
sum_null=train.isnull().sum()
sum_null.index
        
for name, num_na in zip(sum_null.index, train.isnull().sum()):
    if num_na>0:
        print("{}: {}".format(name,num_na))
```
