```python
train = pd.read_csv(r"C:\Users\Ada\1_Jupyter\DMML Lab\Progettino\training_ready.csv")
train.head()
```

## Pattern Mining

```python
# We transform numeric variables into categorical variables (except for VehicleAge which is already in a suitable format)
# We create a second dataframe where we will store the new variables and delete those variables that will not be used in our analysis
```

```python
train2 = train
train2.info()
```

## Features Engineering 

```python
# We create bins that will be the categories of our new variables 
```

```python
train2['VehOdo_bin'] = pd.cut(train2['VehOdo'].astype(int), 10, right=False)
train2['WarrantyCost_bin'] = pd.cut(train2['WarrantyCost'].astype(int), 10, right=False)
train2['AverageMMR__bin'] = pd.cut(train2['AverageMMR'].astype(int), 10, right=False)
train2['VehBCost_bin'] = pd.cut(train2['VehBCost'].astype(int), 10, right=False)

train2.drop(["VehOdo_Standard","OutCost",'VehOdo', "WarrantyCost", "AverageMMR", "VehBCost", "VehBCost"], axis=1, inplace=True)
```

```python
train2.head()
```

```python
train2.info()
```

```python
# Remapping categorical variables: 
# We transform all numerical variables into strings and add labels to ease the analysis and interpretation of results
```

```python
train2['IsBadBuy'] = train2['IsBadBuy'].map(
    {0: 'NotBadBuy', 1: 'IsBadBuy'}).astype(str)
train2['Size_class'] = train2['Size_class'].astype(str) + "_Size"
train2['VehBCost_bin'] = train2['VehBCost_bin'].astype(str) + "_VebCost"
train2['Auction'] = train2['Auction'].astype(str) + "_Auction"

train2['VehOdo_bin'] = train2['VehOdo_bin'].astype(str) + "_VehOdo"
train2['WarrantyCost_bin'] = train2['WarrantyCost_bin'].astype(str) + "_WarrantyCost"
train2['AverageMMR__bin'] = train2['AverageMMR__bin'].astype(str) + "_AverageMMR"
```

```python
train2.head()
```

## Pattern Mining Algorithm

```python
# We create a list that contains the records
```

```python
baskets = train2.values.tolist()
baskets[1]
```

```python
# We extract the association rules
```

```python
rules2 = apriori(baskets, supp=1, zmin=4, zmax= 4, target='r', conf=20, report='ascl') 
```

```python
print('Number of rule:', len(rules2))
```

```python
# We set parameters in order to the filter the rules with appropriate lift and confidence levels.
# r[5] = lift, r[4] = Conf. We set lift>2 and confidence > 0.6
```

```python
count = 0
for r in rules2:
    if r[5] > 2 and r[4] > 0.6:
#         print(r)
        count += 1
print(count)
```

```python
# With lift>2 and confidence > 0.60 we obtained 346597 rules.
# However, none of the rules implicates "NotBadBuy" nor "IsBadBuy", which are the classes of our target variable.
# We play with lift and confidence level in order to obtain valuable rules. 
```

```python
count = 0
for r in rules2:
    if r[0] == "NotBadBuy" and r[5] > 1.1 and r[4] > 0.2:
        print(r)
        count += 1
print(count)
```

```python
# We focus on rules implicating "IsBadBuy" in order to understand which sets of features migh predict a "Bad Buy".
```

```python
count = 0
for r in rules2:
    if r[0] == "IsBadBuy" and r[5] > 2.2 and r[4] > 0.2:
        print(r)
        count += 1
print(count)
```
